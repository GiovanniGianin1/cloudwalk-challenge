1. Objective
The purpose of this report is to analyze hourly transaction data from two checkout points and identify anomaly patterns by comparing today's performance with historical benchmarks. This analysis simulates real-world monitoring tasks and supports the development of alerting systems.

2. Methodology
- Data Sources: checkout_1.csv and checkout_2.csv
- Tools Used: Python (Pandas, Seaborn, Matplotlib), SQLite (in-memory)
- Steps:
- Data cleaning and formatting
- Visual comparison of hourly sales
- Statistical anomaly detection using standard deviation
- SQL query simulation to filter anomalies

3. Key Findings
Checkout 1
- Significant spikes at 10h, 12h, and 15h, where today's sales were much higher than weekly and monthly averages.
- These deviations exceeded 2 standard deviations, indicating potential anomalies.
Checkout 2
- Sudden drop to zero transactions between 15h and 17h, while historical averages were consistently above 20.
- This behavior suggests a possible system failure, POS outage, or data ingestion issue.

4. SQL-Based Validation
A SQL query was executed to identify hours where today's sales were more than twice the weekly average:

    ```sql
        SELECT time, today, avg_last_week
        FROM checkout1
        WHERE today > (avg_last_week * 2)
        ORDER BY time;
    ```


This confirmed the same anomaly hours detected via statistical methods, reinforcing the reliability of the findings.

5. Hypotheses
- Spikes may be caused by promotional campaigns, duplicated transactions, or batch processing delays.
- Drops may result from infrastructure issues, network failures, or POS downtime.

6. Recommendations
- Implement real-time monitoring to detect sudden spikes or drops in transaction volume.
- Set dynamic thresholds based on historical averages and standard deviation.
- Investigate operational logs for the 15h–17h window in Checkout 2.
- Use SQL queries as part of automated anomaly filtering pipelines.

7. Next Steps
This analysis lays the foundation for Task 3.2, where a real-time alerting system will be developed to automatically detect and notify anomalies in transaction behavior.
